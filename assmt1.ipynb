```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.DataFrame()
iris='C:\\Users\HP\Desktop\iris.xls'
df=pd.read_excel(iris)
df.values
df
#Displaying column values
```

```python
#Data Visualisation Type : Sequential Heatmap

df2 = df[['SL','SW','PL','PW']]
print(df2)
plt.figure(figsize=(6,6))
sns.heatmap(df2,annot=True,cmap='Blues')
plt.xlabel('Flower Species')
plt.ylabel('Expected Length/Width')
```

```python
#Data Visualisation Type:Qualitative Heatmap
df2 = df[['SL','SW','PL','PW']]
print(df2)
plt.figure(figsize=(6,6))
sns.heatmap(df2,annot=True,cmap='Pastel1')
plt.xlabel('Flower Species')
plt.ylabel('Expected Length/Width')
```

```python
#Data Visualisation Type:Histogram
df2 = df[['SL','SW','PL','PW']]
print(df2)
df2['SL'].hist(bins=30)
plt.xticks(rotation=90)
plt.xlabel('Expected Sepel Length')
plt.style.use('seaborn')
```

```python
#Data Visualisation Type:Histogram
df2 = df[['SL','SW','PL','PW']]
print(df2)
df2['SW'].hist(bins=30)
plt.xticks(rotation=90)
plt.xlabel('Expected Sepel Width')
plt.style.use('seaborn')
```

```python
#Data Visualisation Type:Subplot
plt.figure(figsize=((15,2)))
pxlcolnames=df2.columns[:-1]
plt.subplot(1,5,1)
image_values=df2.loc[1,pxlcolnames].values
image_label=df2.loc[1,'SW']
plt.imshow(image_values.reshape(3,1),cmap='gray')
plt.title('Sepel Width:'+str(image_label))
```
